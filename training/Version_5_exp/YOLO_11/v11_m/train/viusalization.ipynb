```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = "/home/sur06423/project/Version_5_exp/YOLO_11/v11_m/train/results.csv"  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Define the loss columns and their labels
loss_columns = {
    "box_loss": ("train/box_loss", "val/box_loss"),
    "cls_loss": ("train/cls_loss", "val/cls_loss"),
    "dfl_loss": ("train/dfl_loss", "val/dfl_loss"),
}

# Generate a plot for each type of loss
for loss_name, (train_col, val_col) in loss_columns.items():
    plt.figure(figsize=(10, 6))
    plt.plot(data["epoch"], data[train_col], label=f"Training {loss_name}", linestyle="--", marker="o")
    plt.plot(data["epoch"], data[val_col], label=f"Validation {loss_name}", linestyle="-", marker="x")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"YOLO11m: Training vs Validation {loss_name.capitalize()}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plot learning rates vs epochs
plt.figure(figsize=(10, 6))
plt.plot(data["epoch"], data["lr/pg0"], label="Learning Rate PG0", linestyle="-", marker="o")
plt.plot(data["epoch"], data["lr/pg1"], label="Learning Rate PG1", linestyle="--", marker="x")
plt.plot(data["epoch"], data["lr/pg2"], label="Learning Rate PG2", linestyle=":", marker="s")
plt.xlabel("Epoch")
plt.ylabel("Learning Rate")
plt.title("Learning Rates vs Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate average gap between training and validation box loss
data['box_loss_gap'] = data['val/box_loss'] - data['train/box_loss']
average_box_loss_gap = data['box_loss_gap'].mean()
print(f"The average gap between training and validation box loss over 100 epochs is: {average_box_loss_gap}")
```

```python
# Calculate average gap between training and validation cls loss
data['cls_loss_gap'] = data['val/cls_loss'] - data['train/cls_loss']
average_cls_loss_gap = data['cls_loss_gap'].mean()
print(f"The average gap between training and validation cls loss over 100 epochs is: {average_cls_loss_gap}")
```

```python
# Calculate average gap between training and validation dfl loss
data['dfl_loss_gap'] = data['val/dfl_loss'] - data['train/dfl_loss']
average_dfl_loss_gap = data['dfl_loss_gap'].mean()
print(f"The average gap between training and validation dfl loss over 100 epochs is: {average_dfl_loss_gap}")
```
